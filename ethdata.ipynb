```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
price_gas_corr=df.price.corr(df.gasUsed)
price_value_corr = df.price.corr(df.value)
gas_value = df.gasUsed.corr(df.value)
```

```python
print(f"THE CORRELATION BETWEEN PRICE AND GAS IS {price_gas_corr}")
print(f"THE CORRELATION BETWEEN PRICE AND VALUE IS {price_value_corr}")
print(f"THE CORRELATION BETWEEN GAS AND VALUE IS {gas_value}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df.price, df.gasUsed, color='green', marker='o', s=50)
plt.xlabel('Price')
plt.ylabel('Gas')
plt.title(f'Price Gas Scatter plot with correlation {price_gas_corr} ')
```

```python
plt.scatter(df.price, df.value, color='orange', marker='o', s=50)
plt.xlabel('Price')
plt.ylabel('Value')
plt.title(f'Price Value Scatter plot with correlation {price_value_corr} ')
```

```python
plt.scatter(df.gasUsed, df.value, color='red', marker='o', s=50)
plt.xlabel('Gas')
plt.ylabel('Value')
plt.title(f'Gas Value Scatter plot with correlation {gas_value} ')
```
